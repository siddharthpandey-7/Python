```python
## Class Inheritance
## Create a base class names Person with attributes name and age.
# Create a derived class named Employee that inherits from Person and adds an attribute employee_id.
# Create an object of the derived class and print its attributes.

class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

class Employee(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

employee = Employee('Aman', 30, 'A158')
print(employee.name, employee.age, employee.employee_id)
```

```python
## In the Employee class, override the __str__ method to return a string representation of the object. 
## Create an object of the class and print it.

# Define the parent class Person first
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Define the subclass Employee, which inherits from Person
class Employee(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)  # Call constructor of Person
        self.employee_id = employee_id

    def __str__(self):
        return f"Employee(Name: {self.name}, Age: {self.age}, Employee ID: {self.employee_id})"

# Create an object of Employee
employee = Employee('Siddharth', 20, 'A189')

# Print the object (calls __str__)
print(employee)

```

```python
## Create a class named Counter with a class variable count. 
## Each time an object is created, increment the count. Add a method to get the current count.
# Create multiple objects and print the count

# Define a class named Counter
class Counter:
    # Class variable to keep track of how many objects are created
    count = 0

    # Constructor method called every time a new object is created
    def __init__(self):
        Counter.count += 1  # Increment the class variable 'count' by 1

    # Class method to return the current value of 'count'
    @classmethod
    def get_count(cls):
        return cls.count  # 'cls' refers to the class itself, like 'Counter'

#Test the Counter class 
# Create three instances (objects) of Counter
c1 = Counter()
c2 = Counter()
c3 = Counter()

# Print how many Counter objects have been created so far
print(Counter.get_count())  # Output: 3
```

```python

#Create a class named Rectangle with private attributes length and width. 
#Use properties to get and set these attributes. 
#Create an object of the class and test the properties.

# Define a class named Rectangle
class Rectangle:
    # Constructor to initialize private attributes __length and __width
    def __init__(self, length, width):
        self.__length = length  # Double underscore makes it a private variable
        self.__width = width    # Private variable

    # Getter method for length using @property decorator
    @property
    def length(self):
        return self.__length  # Return the value of private variable __length

    # Setter method for length using @length.setter decorator
    @length.setter
    def length(self, length):
        self.__length = length  # Set a new value for __length

    # Getter method for width
    @property
    def width(self):
        return self.__width  # Return the value of __width

    # Setter method for width
    @width.setter
    def width(self, width):
        self.__width = width  # Set a new value for __width


# Create a Rectangle object with length 10 and width 5
rect = Rectangle(10, 5)

# Access length and width using getter properties
print(rect.length, rect.width)  # Output: 10 5

# Modify length and width using setter properties
rect.length = 15
rect.width = 7

# Check the updated values
print(rect.length, rect.width)  # Output: 15 7
```

```python
##Create an abstract base class named Shape with an abstract method area. 
# Create derived classes Circle and Square that implement the area method. 
# Create objects of the derived classes and call the area method.

# Import ABC (Abstract Base Class) and abstractmethod from the abc module
from abc import ABC, abstractmethod
import math  # Needed for math.pi in the Circle area calculation

# Define an abstract class named Shape
class Shape(ABC):
    # Abstract method that must be implemented by all subclasses
    @abstractmethod
    def area(self):
        pass  # No implementation here – just a placeholder

# Define a subclass Circle that inherits from Shape
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius  # Initialize the radius of the circle

    def area(self):
        # Implement the area formula for a circle: π * r^2
        return math.pi * self.radius ** 2

# Define another subclass Square that inherits from Shape
class Square(Shape):
    def __init__(self, side):
        self.side = side  # Initialize the side length of the square

    def area(self):
        # Implement the area formula for a square: side^2
        return self.side ** 2


# Create a Circle object with radius 5
circle = Circle(5)

# Create a Square object with side length 4
square = Square(4)

# Print the area of the circle
print(circle.area())  

# Print the area of the square
print(square.area())  


```

```python
##Create a class named Vector with attributes x and y. 
# Overload the + operator to add two Vector objects. 
# Create objects of the class and test the operator overloading.

# Define a class named Vector
class Vector:
    # Constructor to initialize x and y coordinates of the vector
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overload the + operator to add two Vector objects
    def __add__(self, other):
        # Returns a new Vector with summed x and y values
        return Vector(self.x + other.x, self.y + other.y)

    # String representation to display the vector nicely when printed
    def __str__(self):
        return f"Vector({self.x}, {self.y})"


# Create the first vector v1 with coordinates (2, 3)
v1 = Vector(2, 3)

# Create the second vector v2 with coordinates (4, 5)
v2 = Vector(4, 5)

# Use the overloaded + operator to add v1 and v2
v3 = v1 + v2  # This calls the __add__ method internally

# Print the result
print(v3)  # Output- Vector(6, 8)
```

```python
## Create a custom exception named InsufficientBalanceError.
#  In the BankAccount class, raise this exception when a withdrawal amount is greater than the balance.
#  Handle the exception and print an appropriate message.

# Define a custom exception class for insufficient balance
class InsufficientBalanceError(Exception):
    pass  # No extra behavior needed, it just inherits from base Exception

# Define a class to simulate a Bank Account
class BankAccount:
    # Constructor to initialize account number and optional initial balance
    def __init__(self, account_number, balance=0):
        self.__account_number = account_number  # Private variable for account number
        self.__balance = balance  # Private variable for balance

    # Method to deposit money into the account
    def deposit(self, amount):
        self.__balance += amount  # Add the amount to the current balance

    # Method to withdraw money from the account
    def withdraw(self, amount):
        # Check if the balance is enough for the withdrawal
        if amount > self.__balance:
            # If not, raise a custom exception
            raise InsufficientBalanceError("Insufficient balance!")
        else:
            self.__balance -= amount  # Deduct the amount from balance if enough

    # Method to check current balance
    def check_balance(self):
        return self.__balance  # Return the private balance value


# Create a BankAccount object with account number '12345678' and initial balance 1000
account = BankAccount('12345678', 1000)

# Deposit 500 into the account
account.deposit(500)  # New balance becomes 1500

# Try to withdraw 2000 (which is more than the available balance)
try:
    account.withdraw(2000)
except InsufficientBalanceError as e:
    print(f"Error: {e}")  # Output: Error: Insufficient balance!

```

```python
#Create a class named FileManager that implements the context manager protocol to open and close a file.
#  Use this class to read the contents of a file.

# Define a class FileManager to manage file operations with context management
class FileManager:
    # Constructor to initialize the filename and mode (e.g., 'r' for read, 'w' for write)
    def __init__(self, filename, mode):
        self.filename = filename  # Store the filename
        self.mode = mode  # Store the mode (read, write, etc.)

    # __enter__ method is used when the 'with' statement is executed
    def __enter__(self):
        # Open the file using the provided filename and mode
        self.file = open(self.filename, self.mode)
        # Return the file object so it can be used in the 'with' block
        return self.file

    # __exit__ method is used when the 'with' block ends (or an exception occurs)
    def __exit__(self, exc_type, exc_value, traceback):
        # Close the file to ensure resources are released after the 'with' block
        self.file.close()


# Use the 'with' statement to automatically open and close the file
# 'r' means read mode
# 'sample.txt' is the file being opened
with FileManager('sample.txt', 'r') as file:
    # Read the content of the file
    content = file.read()
    # Print the content of the file
    print(content)
```

```python
#Create a class named Calculator with methods to add, subtract, multiply, and divide.
#  Each method should return the object itself to allow method chaining. 
# Create an object and chain multiple method calls.

# Define a Calculator class to perform basic arithmetic operations
class Calculator:
    # Constructor to initialize the value (default is 0)
    def __init__(self, value=0):
        self.value = value  # Store the initial value

    # Method to add a number to the current value
    def add(self, amount):
        self.value += amount  # Increment the current value by 'amount'
        return self  # Return the Calculator object to allow method chaining

    # Method to subtract a number from the current value
    def subtract(self, amount):
        self.value -= amount  # Decrease the current value by 'amount'
        return self  # Return the Calculator object to allow method chaining

    # Method to multiply the current value by a number
    def multiply(self, amount):
        self.value *= amount  # Multiply the current value by 'amount'
        return self  # Return the Calculator object to allow method chaining

    # Method to divide the current value by a number
    def divide(self, amount):
        if amount != 0:  # Ensure that division by zero doesn't occur
            self.value /= amount  # Divide the current value by 'amount'
        else:
            print("Cannot divide by zero!")  # Print error message if dividing by zero
        return self  # Return the Calculator object to allow method chaining


# Create a Calculator object with default value 0
calc = Calculator()

# Perform a series of operations using method chaining
# Add 10, subtract 3, multiply by 2, and divide by 2
calc.add(10).subtract(3).multiply(2).divide(2)

# Print the final value after the operations
print(calc.value)  # Output: 7.0
```
